## Preprocess the data

### Subtask:
Clean the email text and extract relevant features (word counts for "account" and "login").

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/emails.csv")

# Drop non-numerical columns (keeping only word counts + spam labels)
df_numeric = df.select_dtypes(include=['int64', 'float64'])
df_numeric['label'] = df['Prediction']  # Add spam label back (0 = ham, 1 = spam)

# Scatter Plot 1: Spam vs Non-Spam Emails (account vs login)
plt.figure(figsize=(8,6))
spam = df_numeric[df_numeric['label'] == 1]
ham = df_numeric[df_numeric['label'] == 0]

plt.scatter(spam['account'], spam['login'], color='red', marker='*', label='Spam Emails')
plt.scatter(ham['account'], ham['login'], color='blue', marker='o', label='Non-Spam Emails')

plt.title("Account & Login Word Frequency in Spam vs Non-Spam Emails")
plt.xlabel("Count of 'account'")
plt.ylabel("Count of 'login'")
plt.legend()
plt.show()

# Scatter Plot 2: Hours Studied vs Test Scores
hours = [2, 3, 4, 5, 6, 7, 8, 9, 10]
scores = [40, 50, 55, 65, 70, 75, 80, 85, 95]

plt.figure(figsize=(8,6))
for h, s in zip(hours, scores):
    if s >= 75:
        plt.scatter(h, s, color='red', marker='^', label='High Score' if 'High Score' not in plt.gca().get_legend_handles_labels()[1] else "")
    else:
        plt.scatter(h, s, color='blue', marker='o', label='Low/Medium Score' if 'Low/Medium Score' not in plt.gca().get_legend_handles_labels()[1] else "")

plt.title("Study Hours vs Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Test Scores")
plt.legend()
plt.show()
```
