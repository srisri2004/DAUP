```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 20   # number of questions
p = 0.25 # probability of correct answer when guessing

# 1. Distribution Characteristics
mean = n * p
variance = n * p * (1 - p)
std_dev = np.sqrt(variance)

print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)

# Mode (most likely number of correct answers)
mode = int((n + 1) * p)
print("Most Likely Correct Answers (Mode):", mode)

# 2. Exact Probability Question
prob_at_least_10 = 1 - binom.cdf(9, n, p)
print("P(X ≥ 10):", prob_at_least_10)

# 3. Performance Evaluation
prob_less_than_5 = binom.cdf(4, n, p)
prob_between_5_and_8 = binom.cdf(8, n, p) - binom.cdf(4, n, p)

print("P(X < 5):", prob_less_than_5)
print("P(5 ≤ X ≤ 8):", prob_between_5_and_8)

# Create dataset for 0–20 correct answers
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
cdf = binom.cdf(x, n, p)

data = pd.DataFrame({
    "Correct_Answers": x,
    "PMF": pmf,
    "CDF": cdf
})

# Save to CSV
csv_file = "binomial_exam_distribution.csv"
data.to_csv(csv_file, index=False)
print(f"Dataset saved as {csv_file}")

# Plot distribution
plt.bar(x, pmf, color='skyblue')
plt.title("Binomial Distribution: Guessing 20 Questions (p=0.25)")
plt.xlabel("Number of Correct Answers")
plt.ylabel("Probability")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
