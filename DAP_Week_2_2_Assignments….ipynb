```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Lead Generation Matrix
L = np.array([
    [200, 150, 180],  # Q1
    [220, 160, 200],  # Q2
    [250, 170, 210]   # Q3
])

# Step 2: Cost-per-Lead Matrix (₹)
C = np.array([
    [50, 40, 60],   # Q1
    [55, 42, 58],   # Q2
    [53, 41, 62]    # Q3
])

# Step 3: Total Marketing Cost (Element-wise multiplication)
T = L * C
print("Total Marketing Cost Matrix (₹):\n", T)

# Step 4: Simulate Growth for Q4 (+15%)
L_Q4 = L * 1.15
print("\nProjected Lead Matrix with Q4 (+15% growth):\n", L_Q4)

# Step 5: Transpose the Growth Matrix
L_proj = np.hstack((L, L_Q4[0:1].T))  # combine Q1–Q3 with Q4 row-wise
L_proj_T = np.vstack((L.T, L_Q4[0]))  # safer method
print("\nTransposed Projected Matrix:\n", L_proj_T)

# Interpretation: Totals
platforms = ["Social Media", "Email", "Paid Ads"]
quarters = ["Q1", "Q2", "Q3"]

total_leads = L.sum(axis=0)
total_costs = T.sum(axis=0)
cost_per_lead = total_costs / total_leads

print("\nTotal Leads per Platform (Q1–Q3):", dict(zip(platforms, total_leads)))
print("Total Costs per Platform (Q1–Q3):", dict(zip(platforms, total_costs)))
print("Average Cost per Lead (Q1–Q3):", dict(zip(platforms, cost_per_lead)))

best_efficiency = platforms[np.argmin(cost_per_lead)]
print("\nBest Cost Efficiency Platform:", best_efficiency)

# ================= Visualization =================

# Leads per platform (Q1–Q3)
plt.figure(figsize=(8,5))
for i, platform in enumerate(platforms):
    plt.plot(quarters, L[:,i], marker='o', label=platform)
plt.title("Leads Generated (Q1–Q3)")
plt.xlabel("Quarter")
plt.ylabel("Leads")
plt.legend()
plt.show()

# Cost efficiency bar chart
plt.figure(figsize=(8,5))
plt.bar(platforms, cost_per_lead, color=['#4CAF50','#2196F3','#FF9800'])
plt.title("Average Cost per Lead (Q1–Q3)")
plt.ylabel("₹ per Lead")
plt.show()
```
