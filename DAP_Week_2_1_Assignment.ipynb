```python
# Step 0: Install required libraries (NumPy + Matplotlib for visualization)
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Revenue Matrix (₹ in lakhs)
R = np.array([
    [500, 300, 200],  # Store A
    [450, 250, 300],  # Store B
    [600, 350, 250]   # Store C
])

# Step 2: Expenses Matrix (₹ in lakhs)
E = np.array([
    [300, 200, 150],  # Store A
    [280, 180, 200],  # Store B
    [350, 220, 180]   # Store C
])

# Step 3: Net Revenue = Revenue - Expenses
N = R - E
print("Net Revenue Matrix (₹ lakhs):\n", N)

# Step 4: Adjust for Inflation (8%)
R_adj = R * 1.08
print("\nInflation Adjusted Revenue Matrix (₹ lakhs):\n", R_adj)

# Step 5: Transpose the Adjusted Matrix
R_adj_T = R_adj.T
print("\nTransposed Inflation Adjusted Revenue Matrix (₹ lakhs):\n", R_adj_T)

# Step 6: Interpretation
stores = ["Store A", "Store B", "Store C"]
categories = ["Groceries", "Clothing", "Electronics"]

# Store totals (Net Revenue)
store_totals = N.sum(axis=1)
# Category totals (Net Revenue)
category_totals = N.sum(axis=0)

print("\nStore-wise Net Revenue Totals (₹ lakhs):", dict(zip(stores, store_totals)))
print("Category-wise Net Revenue Totals (₹ lakhs):", dict(zip(categories, category_totals)))

# Best performing store
best_store_index = np.argmax(store_totals)
print("\nMost Profitable Store:", stores[best_store_index])

# Best performing category
best_category_index = np.argmax(category_totals)
print("Best Performing Category Across Stores:", categories[best_category_index])

# Best store-category combination after inflation adjustment
max_value = np.max(R_adj_T)
row, col = np.where(R_adj_T == max_value)
print("Best Store-Category Combination (after inflation):",
      categories[row[0]], "in", stores[col[0]], "with ₹", max_value, "lakhs")

# ---------- Visualization ----------
# 1. Bar Chart: Net Revenue per Store
plt.figure(figsize=(8,5))
plt.bar(stores, store_totals, color=['#4CAF50','#2196F3','#FF9800'])
plt.title("Net Revenue by Store (₹ lakhs)")
plt.ylabel("Net Revenue (₹ lakhs)")
plt.show()

# 2. Bar Chart: Net Revenue per Category
plt.figure(figsize=(8,5))
plt.bar(categories, category_totals, color=['#8BC34A','#03A9F4','#FF5722'])
plt.title("Net Revenue by Category (₹ lakhs)")
plt.ylabel("Net Revenue (₹ lakhs)")
plt.show()

# 3. Heatmap-like visualization of Net Revenue Matrix
plt.figure(figsize=(6,5))
plt.imshow(N, cmap="YlGnBu", aspect="auto")
plt.colorbar(label="Net Revenue (₹ lakhs)")
plt.xticks(np.arange(len(categories)), categories)
plt.yticks(np.arange(len(stores)), stores)
plt.title("Net Revenue Matrix Heatmap")
plt.show()
```
